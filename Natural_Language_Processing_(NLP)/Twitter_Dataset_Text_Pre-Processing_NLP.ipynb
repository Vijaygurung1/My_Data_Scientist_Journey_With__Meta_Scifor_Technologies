- As in my outcome there is a [ # & @ ] symbols. So for that I will 1st remove those noises & keep only the Text.

- There is function called - [ remove_punctuation ] 

- what string.punctuation actually is all these noises symbols. I have used that to remove all the symbols and keep text only.

```python
import string 
string.punctuation

def remove_punctuation(text):
    return ''.join([i for i in text if i not in string.punctuation])
```

```python
string.punctuation
```

```python
# Here I have create new column called - 'tweet_clean', where we can see orginal Tweet its clear all the symbols.
```

```python
df['tweet_clean'] = df['tweet'].apply(lambda x: remove_punctuation(x))
df.head()
```

```python
# Here another function it splits the text into words.
```

```python
import re 
def tokenize(text):
    return text.split(' ')

df['tweet_tokenized'] = df['tweet_clean'].apply(lambda x: tokenize(x))
df.head()
```

```python
from nltk.stem.porter import PorterStemmer 

stemmer = PorterStemmer() 
def stemm(text):
    return [stemmer.stem(word) for word in text] 

df['tweet_stemmed'] = df['tweet_tokenized'].apply(lambda x: stemm(x))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
nltk.download("omw-1.4")
# Initialize wordnet lemmatizer
wnl = WordNetLemmatizer()
def lemmatization(text):
    return [wnl.lemmatize(w, pos='v') for w in text]
```

```python
def lemmatization(text):
    return [wnl.lemmatize(w, pos='v') for w in text]

df['tweet_lemmatized'] = df['tweet_tokenized'].apply(lambda x: lemmatization(x))
```
