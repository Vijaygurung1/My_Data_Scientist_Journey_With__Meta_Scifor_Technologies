```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = {
    'Feature1': [1, 2, 3, 4, 5],
    'Feature2': [2, 4, 6, 8, 10],
    'Target': [3, 7, 5, 11, 13]
}

# Load data into a DataFrame
df = pd.DataFrame(data)
X = df[['Feature1', 'Feature2']].values
y = df['Target'].values
```

```python
#2. Implement Multiple Linear Regression from Scratch
```

```python
class MultipleLinearRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
    
    def fit(self, X, y):
        # Number of training examples and features
        self.m, self.n = X.shape
        
        # Initialize the parameters (weights)
        self.beta = np.zeros(self.n + 1)
        
        # Add a column of ones to X for the intercept term
        X = np.c_[np.ones((self.m, 1)), X]
        
        # Gradient Descent
        for _ in range(self.iterations):
            self.beta -= self.learning_rate * self.gradient(X, y)
        
        return self
    
    def gradient(self, X, y):
        y_pred = X.dot(self.beta)
        error = y_pred - y
        grad = X.T.dot(error) / self.m
        return grad
    
    def predict(self, X):
        # Add a column of ones to X for the intercept term
        X = np.c_[np.ones((X.shape[0], 1)), X]
        return X.dot(self.beta)

# Create and fit the model
model = MultipleLinearRegression(learning_rate=0.01, iterations=1000)
model.fit(X, y)

# Predict
predictions = model.predict(X)
print("Predictions:", predictions)
print("Parameters:", model.beta)
```

```python
#3. Visualize the Results Using Matplotlib
```

```python
# Visualization
plt.figure(figsize=(10, 6))

# Plot actual values
plt.scatter(df['Feature1'], y, color='red', label='Actual')

# Plot predicted values
plt.scatter(df['Feature1'], predictions, color='blue', label='Predicted')

# Adding labels and title
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.title('Multiple Linear Regression')
plt.legend()

# Show plot
plt.show()
```
